```python
import matplotlib.pyplot as plt #FOR GRAPH
```

```python
temp = np.array([0, 15, 20, 35, 67])
length = np.array([90, 95, 112, 154, 300])
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(temp, length, color = "red", marker = "*")
plt.xlabel("Temperature in celcius")
plt.ylabel("Length of the stick")
plt.title("Relationship Between T and L")
plt.grid(alpha = 0.2) # For the darkness the grid lines
plt.savefig("Reln_between_t_and_l.png", dpi = 300)#.png, dpi = 300" or .svg for clearer picture
plt.show()
```

```python
plt.plot(temp, length)
plt.show()
```

```python
temp = np.array([0, 15, 20, 35, 67])
length_of_copper = np.array([90, 95, 112, 154, 300])
length_of_aluminium = np.array ([85, 70, 102, 250, 290])
```

```python
plt.scatter(temp, length_of_copper, marker = "*", color = "red", label = "Length of copper" )
plt.scatter(temp,  length_of_aluminium, color = "blue", label = "Length of Aluminium" )
plt.xlabel("Temperature in celcius")
plt.ylabel("Length")
plt.grid(alpha = 0.2)
plt.legend() #for the labels to show
plt.show()
```
